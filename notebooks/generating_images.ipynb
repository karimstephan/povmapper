```python
import requests
N = len(df)
key=
size = '1024x1024'
zoom = '16'
maptype = 'satellite'
path = '/Volumes/GoogleDrive/My Drive/povertymapper1024/'



for i in range(6240, 10000, 1):
    lat = df.lat_cen[i]
    lon = df.lon_cen[i]
    label = df.cats[i]
    url = f'https://maps.googleapis.com/maps/api/staticmap?center={lat},{lon}&zoom={zoom}&size={size}&key={key}&maptype={maptype}'
    image = requests.get(url).content
    print(i)
    print(url)
    f = open(path + '/' + str(label) + '/' + str(i) + '.png', 'wb')
    f.write(image)
    f.close()
```
