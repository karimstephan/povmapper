```python
import pandas as pd
print(f'Pandas: {pd.__version__}')
import numpy as np
print(f'Numpy: {np.__version__}')
import matplotlib.pyplot as plt
import os
from skimage.transform import resize
%matplotlib inline
```

```python
csv_path = '/content/drive/My Drive/df_main_new.csv'
file_path = '/content/drive/MyDrive/Combined_4km_256x256'
y_category = 'cats'

start_row = 1
end_row = 4

df = pd.read_csv(csv_path)
print(f'Shape:{str(df.shape)}')

X = []
y = []

for index, row in df.head().iterrows():
  if index >= start_row:
    if index < end_row:
      print('Adding Image '+str(row['hv001']))
      img_array = np.load(file_path+'/'+str(row['hv001'])+'.npy')
      X.append(img_array)
      y.append(row[y_category])

print(f'{str(len(X))} Images Appended')
print(f'{str(len(y))} Observations Appended')
X = np.array(X)
y = np.array(y)
print('X Array Shape '+str(X.shape))
print('y Array Shape '+str(y.shape))
```
